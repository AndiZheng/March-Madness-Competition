```python
import random
fileName = r'C:\Users\AndyZ\OneDrive\Desktop\March Madness\NCAA.csv'
newfile = r'C:\Users\AndyZ\OneDrive\Desktop\March Madness\shuffle.csv'
with open(fileName, 'r', encoding = 'utf-8') as f1, open(newfile,'w', encoding = 'utf-8') as f2:
    f2.write(s1+'\n')
    line = f1.readline().replace('\n','')
    lineList = line.split(',')
    team1 = []
    team2 = []
    for name in lineList:
        if name.startswith('team1'):
            i1 = lineList.index(name)
            team1.append(i1)
        if name.startswith('team2'):
            i2 = lineList.index(name)
            team2.append(i2)

    line = f1.readline().replace('\n','')
    while line != '':
        randomInt = random.randint(0,1)
        if randomInt == 1:
            f2.write(line+'\n')           
        if randomInt == 0:
            l1 = line.split(',')
            length = len(l1)
            for data in range(0,length):
                dataIndex = data
                if dataIndex in team1:
                    team1DataIndex = dataIndex
                    team2DataIndex = team2[team1.index(dataIndex)]
                    l1[team1DataIndex],l1[team2DataIndex] = l1[team2DataIndex],l1[team1DataIndex]
                else:
                    pass
            newstr = ','.join(l1)
            f2.write(newstr+'\n')
        line = f1.readline().replace('\n','')
                                           
                                           
                    
            


            
 
            
            
    
        
```
