```python
import pandas as pd
import numpy as np
df = pd.read_csv('shuffle.csv')
df.head()
```

```python
game_id = df['game_id'].tolist()
df = df.drop(['game_id'],axis = 1)
df.head()
```

```python
array = df.values
x = array[:,0:66]
y = array[:,66]
x.shape, y.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn import cross_validation
sc = StandardScaler()
x_std = sc.fit_transform(x)
pca = PCA(n_components = 'mle',svd_solver='full')
x_pca = pca.fit_transform(x_std)
x_train_pca, x_test_pca, y_train, y_test = cross_validation.train_test_split(x_pca,y,test_size=0.3,random_state = 42)
x_train_pca.shape,x_test_pca.shape
```

```python
# from sklearn.ensemble import RandomForestClassifier, VotingClassifier
# clf1 = LogisticRegression(solver='lbfgs', multi_class='multinomial',
#                           random_state=1)
# clf2 = RandomForestClassifier(n_estimators=50, random_state=1)
# clf3 = GaussianNB()
# eclf1 = VotingClassifier(estimators=[
#         ('lr', clf1), ('rf', clf2), ('gnb', clf3)], voting='hard')
# eclf1 = eclf1.fit(x_pca, y)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
param_test1 = {'n_estimators':range(20,81,10)}
gsearch1 = GridSearchCV(estimator = GradientBoostingClassifier(learning_rate=0.1, min_samples_split=300,
                                  min_samples_leaf=20,max_depth=8,max_features='sqrt', subsample=0.8,random_state=10), 
                       param_grid = param_test1, scoring='roc_auc',iid=False,cv=5)
gsearch1.fit(x_train_pca,y_train)
gsearch1.grid_scores_, gsearch1.best_params_, gsearch1.best_score_
```

```python
param_test2 = {'max_depth':range(3,14,2), 'min_samples_split':range(100,801,200)}
gsearch2_gbc = GridSearchCV(estimator = GradientBoostingClassifier(learning_rate=0.1, n_estimators=80, min_samples_leaf=20, 
      max_features='sqrt', subsample=0.8, random_state=10), 
   param_grid = param_test2, scoring='roc_auc',iid=False, cv=5)
gsearch2_gbc.fit(x_train_pca,y_train)
gsearch2_gbc.grid_scores_, gsearch2_gbc.best_params_, gsearch2_gbc.best_score_
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

RF = RandomForestClassifier()
param_gridRF = [{
                'n_estimators':[50,100,200,700],
                #'learning_rate':[1,2,3],
               #'algorithm':['SAMME','SAMME.R'],
                }]
               

search_RF = GridSearchCV(RF, param_gridRF, cv=10)
search_RF.fit(x_train_pca, y_train)
print("Best parameter (CV score=%0.3f):" % search_RF.best_score_)
print(search_RF.best_params_)
```

```python
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
# clf1 = LogisticRegression(solver='lbfgs', multi_class='multinomial',
#                           random_state=1)
# clf2 = RandomForestClassifier(n_estimators=50, random_state=1)
# clf3 = GaussianNB()
eclf1 = VotingClassifier(estimators=[
        ('gbc', gsearch2_gbc), ('rf', search_RF)], voting='soft')
eclf1 = eclf1.fit(x_train_pca, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = eclf1.predict(x_pca)
accuracy = accuracy_score(y, y_pred)
accuracy
```

```python
y_prob = eclf1.predict_proba(x_pca)
y_prob
```

```python
from sklearn.metrics import log_loss
log_loss(y,y_prob)
```

```python
df['team1_outcome_pred'] = y_pred[:]
team1_win = []
for lst in y_prob:
    p = lst[1]
    team1_win.append(p)
team1_win
```

```python
df['game_id'] = game_id
df['probability'] = team1_win
df.head()
```

```python
outcome = df.drop(df.columns[0:68],axis = 1)
outcome
```

```python
outcome.to_csv('result.csv')
```
